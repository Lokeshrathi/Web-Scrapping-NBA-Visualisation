```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print('Successfully Done')
```

```python
# The Building Blocks
year = '2019'
str = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'

# Combining the URL + year strings together
url = str.format(year)
url
```

```python
years = [2015,2016,2017,2018,2019]
str = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'

for year in years:
  url = str.format(year)
  print(url)
```

```python
df = pd.read_html(url, header = 0)
df
```

```python
#Since there are 2 tables, let us work with the first one
df[0]
```

## Data Cleaning

```python
df2019 = df[0]
```

```python
df2019[df2019.Age == 'Age'] # we see that the table header is presented multiple times in severl rows
```

```python
len(df2019[df2019.Age == 'Age'])
df = df2019.drop(df2019[df2019.Age == 'Age'].index)
df.shape
```

```python
df['PTS'] =df['PTS'].astype('float')
```

```python
sns.distplot(df.PTS,
             kde=False,
             hist_kws=dict(edgecolor="black", linewidth=2),
             color='#00BFC4')
```

```python
df['AST']=df['AST'].astype('float')
sns.distplot(df.AST,
             kde=False,
             hist_kws=dict(edgecolor="black", linewidth=2),
             color='#00BFC4')
```

```python
df['3P'] = df['3P'].astype('float')
sns.distplot(df['3P'],
             kde=False,
             hist_kws=dict(edgecolor="black", linewidth=2),
             color='#00BFC4')
```

```python
top_activities = df.groupby(by='Player')['PTS'].max().sort_values(ascending =False).head(20).reset_index()
plt.figure(figsize=(15,10))
plt.xticks(fontsize =15)
plt.yticks(fontsize=15)
plt.xlabel('Total points' ,fontsize = 15)
plt.ylabel('Player_name' ,fontsize = 15)
plt.title('Top 20 players with most points' ,fontsize=15)

ax = sns.barplot(x=top_activities['PTS'], y = top_activities['Player'])
for i ,(value,name) in enumerate (zip(top_activities['PTS'],top_activities['Player'])):
    ax.text(value, i-.05,f'{value:,.0f}',size = 10,ha='left',va='center')
ax.set(xlabel='Points',ylabel='Player_name')
```

```python
df['AST']=df['AST'].astype('float')
top_assists = df.groupby(by='Player')['AST'].max().sort_values(ascending =False).head(20).reset_index()
plt.figure(figsize=(15,10))
plt.xticks(fontsize =15)
plt.yticks(fontsize=15)
plt.xlabel('Total assists' ,fontsize = 15)
plt.ylabel('Player_name' ,fontsize = 15)
plt.title('Top 20 players with most assists' ,fontsize=15)

ax = sns.barplot(x=top_assists['AST'], y = top_assists['Player'])
for i ,(value,name) in enumerate (zip(top_assists['AST'],top_assists['Player'])):
    ax.text(value, i-.05,f'{value:,.0f}',size = 10,ha='left',va='center')
ax.set(xlabel='Assists',ylabel='Player_name')
```

```python
df['3P'] = df['3P'].astype('float')
top_3pointers = df.groupby(by='Player')['3P'].max().sort_values(ascending =False).head(20).reset_index()
plt.figure(figsize=(15,10))
plt.xticks(fontsize =15)
plt.yticks(fontsize=15)
plt.xlabel('3_pointers' ,fontsize = 15)
plt.ylabel('Player_name' ,fontsize = 15)
plt.title('Top 20 players with most 3 pointers made' ,fontsize=15)

ax = sns.barplot(x=top_3pointers['3P'], y = top_3pointers['Player'])
for i ,(value,name) in enumerate (zip(top_3pointers['3P'],top_3pointers['Player'])):
    ax.text(value, i-.05,f'{value:,.0f}',size = 10,ha='left',va='center')
ax.set(xlabel='3_pointers',ylabel='Player_name')
```
